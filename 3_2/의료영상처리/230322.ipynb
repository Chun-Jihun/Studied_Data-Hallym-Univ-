# 히스토그램

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('Parthenon.jpg',cv2.IMREAD_GRAYSCALE)
#calcHist(입력 이미지배열, 히스토그램을 얻을 채널 인덱스, Mak이미지, X축 요소의 개수, Y축 요소값의 범위로 하나의 채널에 대한 화소 강도가 0~255이므로 대부분 0,256임)
hist = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
```

# 히스토그램 산술연산

```python
imgAdd = cv2.add(img,50)
imgSub = cv2.subtract(img,50)
imgMul = cv2.multiply(img,1.5)
imgDiv = cv2.divide(img,1.5)

histAdd = cv2.calcHist([imgAdd],[0],None,[256],[0,256])
histSub = cv2.calcHist([imgSub],[0],None,[256],[0,256])
histMul = cv2.calcHist([imgMul],[0],None,[256],[0,256])
histDiv = cv2.calcHist([imgDiv],[0],None,[256],[0,256])
```

```python
plt.plot(histAdd)
```

```python
plt.plot(histSub)
```

```python
plt.plot(histMul)
```

```python
plt.plot(histDiv)
```

# 히스토그램 스트레치

```python
#전체적으로 밝아짐
gmin, gmax, _, _ = cv2.minMaxLoc(img)
dst = cv2.normalize(img,None,0,255,cv2.NORM_MINMAX)

cv2.imshow('img',img)
cv2.imshow('dst',dst)
cv2.waitKey()
```
